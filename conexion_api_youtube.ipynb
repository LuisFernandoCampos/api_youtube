```python
api_service_name = "youtube"
api_version = "v3"

youtube = build(api_service_name, api_version, developerKey=api_key)

request = youtube.channels().list(
    part="snippet,statistics,contentDetails",
    id=channel_ids
)

response = request.execute()
```

```python
# Extraer la clave "items"
youtubers_df = pd.DataFrame(response["items"])

# Imprimir el DataFrame
youtubers_df.head(1)
```

```python
# Obtener los valores de la clave específica en las columnas
youtuber_id = youtubers_df["id"]
channel_name = youtubers_df["snippet"].apply(lambda x: x["title"])
publishedAt = youtubers_df["snippet"].apply(lambda x: x["publishedAt"])
image = youtubers_df["snippet"].apply(lambda x: x["thumbnails"]["high"]["url"])
viewCount = youtubers_df["statistics"].apply(lambda x: x["viewCount"])
subscriberCount = youtubers_df["statistics"].apply(lambda x: x["subscriberCount"])
videoCount = youtubers_df["statistics"].apply(lambda x: x["videoCount"])


# Crear un nuevo DataFrame con los valores obtenidos
influencers_df = pd.DataFrame({
        "youtuber_id": youtuber_id,
        "name": channel_name,
        "published_at": publishedAt,
        "images":image,
        "view_count": viewCount,
        "subscriber_count": subscriberCount,
        "video_count": videoCount
})

# Influencers sin canal de youtube
sin_canal = [
    {"youtuber_id": "", "name": "LA CHOLITA LU", "published_at": "", "images": "https://cdn.forbes.pe/2023/10/la-cholita-lu.jpg", "view_count": "", "subscriber_count": "", "video_count": ""},
    {"youtuber_id": "", "name": "MATEOGLG", "published_at": "", "images": "https://cdn.forbes.pe/2023/10/garrido-lecca-rrss.jpg", "view_count": "", "subscriber_count": "", "video_count": ""}
]

# Agregar los nuevos registros al DataFrame
influencers_df = influencers_df.append(sin_canal, ignore_index=True)

# Imprimir el nuevo DataFrame
influencers_df
```
